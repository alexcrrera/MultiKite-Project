```python
from vessel import Vessel
import gfwapiclient as gfw
import asyncio
import json

import folium
import os
```

```python


def getTokenData(name):
    # assume project root contains "build"
    baseDir = os.getcwd()

    while not os.path.exists(os.path.join(baseDir, "build")):
        parent = os.path.dirname(baseDir)
        if parent == baseDir:
            raise FileNotFoundError("Could not find project root")
        baseDir = parent

    tokenPath = os.path.join(baseDir, "build", "source", "token.json")

    with open(tokenPath, "r") as f:
        return json.load(f)[name]
print("OK LOADED")

vessel1 = Vessel( mmsi=None, lat=None, lon=None, timestamp=None, message="test ok")
print(vessel1.message)
```

```python
# ── Client ────────────────────────────────────────────────────
def buildGFW():
    return gfw.Client(access_token=getTokenData("Global_Fishing_Watch"))

```

# ── Geometry (Cabrera) 

```python
def getCabreraGeometry():
    return {
        "type": "Polygon",
        "coordinates": [[
            [2.85, 39.05],
            [3.10, 39.05],
            [3.10, 39.25],
            [2.85, 39.25],
            [2.85, 39.05]
        ]]
    }
```

# ── Fetch Fishing Events 

```python
async def getFishingEventsInCabrera(client):
    geometry = getCabreraGeometry()

    result = await client.events.get_all_events(
        datasets=["public-global-fishing-events:latest"],
        geometry=geometry,
        start_date="2024-01-01",
        end_date="2024-12-31",
        limit=50,
    )

    events = result.data() if result else []

    vessels = {}

    for e in events:
        vessel = getattr(e, "vessel", None)
        position = getattr(e, "position", None)

        if not vessel or not hasattr(vessel, "id"):
            continue

        vid = vessel.id

        vessels[vid] = {
            "lat": getattr(position, "lat", None) if position else None,
            "lon": getattr(position, "lon", None) if position else None,
            "timestamp": getattr(e, "start", None),
            "message": e  # ← SAME PRINCIPLE
        }

    return vessels
```

# ── Fetch Identity 


```python
async def getVesselIdentity(client, vessel_id):
    result = await client.vessels.get_vessel_by_id(
        id=vessel_id,
        dataset="public-global-vessel-identity:latest",
    )
    return result.data() if result else None
```

```python
async def main():
    client = buildGFW()

    vesselData = await getFishingEventsInCabrera(client)

    vesselsDict = {}

    for vid, pos in vesselData.items():
        gfwVessel = await getVesselIdentity(client, vid)

        if not gfwVessel:
            continue

        v = Vessel(lat=pos["lat"],
        lon=pos["lon"],
        timestamp=pos["timestamp"],
        message=pos["message"]   # ← THIS LINE
    )

        v.loadFromGFW(gfwVessel)

        if v.mmsi:
            vesselsDict[v.mmsi] = v

    return vesselsDict
```

```python
vessels = await main()
```

```python
for vessel in vessels.values():
    print(vessel.message)
```

```python
def plotVessels(vessels):
    # Center on Cabrera
    m = folium.Map(location=[39.15, 2.95], zoom_start=10)

    for v in vessels.values():
        if v.lat is None or v.lon is None:
            continue

        popupText = f"""
        MMSI: {v.mmsi}<br>
        Name: {v.name}<br>
        Flag: {v.flag}<br>
        Type: {v.type}<br>
        """

        folium.Marker(
            location=[v.lat, v.lon],
            popup=popupText,
            tooltip=v.name
        ).add_to(m)

    return m
```

```python
m = plotVessels(vessels)
m
```
